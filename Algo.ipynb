* SY's Trial & Error

````python
def Sort(list: List) -> List:
  sort_list: List = []

  for i in range( len(list) ) :
    while i < len(list):
      if list[i] >= list[i+1]:
        sort_list.append( list[i+1] )
      else:
        sort_list.append( list[i] )

  return sort_list
````

* Solution

  - 주어진 데이터 중 최소값 탐색

  - 최소값을 맨 앞으로 이동

##### Recursive Function(재귀함수) | 자신의 로직을 내부적으로 반복하다가, 일정한 조건을 만족하면 함수를 이탈하여 결과를 도출하는 함수 | 팩토리얼을 통한 재귀함수의 표현



* 재귀함수를 호출하는 부분을 상단에 두고 표현

```
  def function(input):
    if input > value:
      # 입력값이 일정값 이상일 경우

        # 입력값보다 하나 작은 값 반환

```

````python
def factorial1(numeric: int) -> int:
  if numeric > 1:
    return numeric * factorial1(numeric - 1)
  else:
    return numeric
````

````python
start_time_entity = time.time()

print( factorial1(5) )

end_time_entity = time.time()
````
